```python
# Using columns with numerical values to calculate the z-score for each point
numericals = ['exch_usd', 'inflation_annual_cpi', 'gdp_weighted_default']

z_scores = pd.DataFrame()
for column in numericals:
    z_scores[column] = (africa[column] - africa[column].mean()) / africa[column].std()

# Keeping a threshold for the outlier and identifying the outliers
outlier_threshold = 3
outliers = africa[np.abs(z_scores) > outlier_threshold]

# Printing the number of outliers
print("Number of outliers:", len(outliers))

# Plotting outliers and non-outliers. Customizing the colors, background, and removing the grids
marker_size = 5
transparency = 0.7
line_width = 0.5

color_non_outliers = 'blue'
color_outliers = 'red'
plt.figure(figsize=(6, 6))

for i, column in enumerate(numericals):
    ax = plt.subplot(len(numericals), 1, i+1)
    
    plt.scatter(africa.index, africa[column], label='Data', color=color_non_outliers,
                s=marker_size, alpha=transparency, linewidths=line_width)
    plt.scatter(outliers.index, outliers[column], color=color_outliers,
                label='Outliers', s=marker_size, alpha=transparency, linewidths=line_width)
    plt.ylabel(column)
    plt.title(f'{column} with Outliers')
    plt.legend()

    if column == 'gdp_weighted_default':
        ax.grid(False)

plt.xlabel('Index')
plt.tight_layout()
plt.show()
```
